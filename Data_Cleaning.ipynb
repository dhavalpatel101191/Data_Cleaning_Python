```python
import pandas as pd
df = pd.read_csv('../Sample Data/Machine_Data.csv')
```

```python
df
```

```python
drop_dp = df.drop(columns= ['Unnamed: 9','Unnamed: 10'])
drop_dp.head()
```

```python
drop_NaN = drop_dp[drop_dp['Row Type'].notna()]
        
```

```python
drop_NaN.head()
```

```python
column_value = []
counter = 0

for i in drop_NaN['Row Type']:
    if 'first name' in i:
        counter += 1
    column_value.append(counter)
```

```python
drop_NaN['New'] = column_value
```

```python
drop_NaN.head()
```

```python
drop_Extra = drop_NaN[drop_NaN['Row Type'] != 'Row Type']
```

```python
drop_Extra.head()
```

```python
Name_Data = drop_Extra[drop_Extra['Row Type'].str.contains('first name')]
Name_Data.head()
```

```python
drop_dp = df.drop(columns= ['Unnamed: 9','Unnamed: 10'])
drop_dp.head()
```

```python
Name_Data.drop(columns= ['Speed1','Speed2','Electricity','Effort','Weight','Torque'], inplace=True)
```

```python
Name_Data.head()
```

```python
Name_Data['Row Type'] = Name_Data['Row Type'].str[12:]
Name_Data['Iter Number'] = Name_Data['Iter Number'].str[11:]
Name_Data['Power1'] = Name_Data['Power1'].str[5:]
```

```python
Name_Data.head()
```

```python
Name_Data.rename(columns= {'Row Type': 'First Name', 'Iter Number':'Last Name', 'Power1':'Time'}, inplace=True)
```

```python
Name_Data.head()
drop_Extra.head()
```

```python
No_Name_Data = drop_Extra[~drop_Extra['Row Type'].str.contains('first name')]
```

```python
No_Name_Data.head()
```

```python
Semi_Final = pd.merge(left=Name_Data,right=No_Name_Data,how='inner',on='New')
```

```python
Semi_Final.head()
```

```python
Semi_Final.drop(columns= ['New'],inplace=True)
```

```python
Semi_Final
```

```python
Semi_Final.to_csv('Macine_Clean_Data.csv',index=False)
```
